## 2. Statistik tentang data

```python
# Mean, Median, Modus, Etc.
# print(md)
data_internet.groupby(['nama_provinsi']).median()['ruta_berinternet']
```

```python
ms  = data_internet.ruta_berinternet.mode()
print(ms)
```

```python
# dev = data_internet.ruta_berinternet.std()
# print(dev)
data_internet.groupby(['nama_provinsi']).std()['ruta_berinternet']
```

```python
# max = data_internet.ruta_berinternet.max()
# min = data_internet.ruta_berinternet.min()
# print(min, max)
max = data_internet.groupby(['nama_provinsi']).max()['ruta_berinternet']
mean = data_internet.groupby(['nama_provinsi']).mean()['ruta_berinternet']
min = data_internet.groupby(['nama_provinsi']).min()['ruta_berinternet']
tot = data_internet.groupby(['nama_provinsi']).sum()['ruta_berinternet']
tbl = pd.concat([min, mean, max, tot], axis=1)
tbl.columns = ['ruta_berinternet_min', 'ruta_berinternet_mean', 'ruta_berinternet_max', 'ruta_berinternet_sum']
tbl
```

```python
data_internet.groupby(['nama_provinsi', 'nama_kabkota']).sum()['ruta_berinternet']
```

## 3. Visualisasi sederhana

```python
#Tampilkan yang menarik dari data tersebut
ax = data_internet.groupby('nama_provinsi').sum()['ruta_berinternet']
ax = ax.plot(kind='pie', title ="Chart 3.1: Pie Chart", figsize=(20,10), legend=False, fontsize=12, autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2)
ax.set_xlabel("", fontsize=12)
ax.set_ylabel("", fontsize=12)
```

```python
ax = data_internet.sort_values(by=['ruta_berinternet'], ascending=True, )[:10].reset_index()
ax[['nama_provinsi', 'nama_kabkota', 'ruta_berinternet']]
```

```python
ax = ax[['nama_kabkota', 'ruta_berinternet']]
ax = ax.plot('nama_kabkota', kind='bar', title ="Chart 3.2: Top 10 Kab/Kota dengan Ruta Berinternet Terkecil", figsize=(20,10), legend=True, fontsize=12)
ax.set_xlabel("Nama Kab/Kota", fontsize=12)
ax.set_ylabel("Jumlah Ruta Internet", fontsize=12)
```

```python
ax = data_internet.sort_values(by=['ruta_berinternet'], ascending=False)[:10].reset_index()
ax[['nama_provinsi', 'nama_kabkota', 'ruta_berinternet']]
```

```python
ax = ax[['nama_kabkota', 'ruta_berinternet']]
ax = ax.plot('nama_kabkota', kind='bar', title ="Chart 3.3: Top 10 Kab/Kota dengan Ruta Berinternet Terbanyak", figsize=(20,10), legend=True, fontsize=12)
ax.set_xlabel("Nama Provinsi", fontsize=12)
ax.set_ylabel("Jumlah Ruta Internet", fontsize=12)
```

```python
ab = data_internet.groupby('nama_provinsi').sum()['ruta_berinternet']
ax = ab.sort_values(ascending=False)[:10]
# ax.plot.bar()
ax = ax.plot(kind='bar', title ="Chart 3.4: Top 10 Provinsi dengan Ruta Berinternet Terbanyak", figsize=(20,10), legend=True, fontsize=12)
ax.set_xlabel("Nama Provinsi", fontsize=12)
ax.set_ylabel("Jumlah Ruta Internet", fontsize=12)
```
